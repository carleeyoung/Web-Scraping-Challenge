#### NASA Mars News

```python
# URL of page to be scraped
url = 'https://mars.nasa.gov/news/'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
#Scrape website for article title and content
news_story = soup.find('div', class_='slide')


news_title = news_story.find('div', class_='content_title').find('a').text

news_content = news_story.find('div', class_='rollover_description_inner').text
print(news_story)
print(news_title)
print(news_content)
```

#### Mars Hemispheres

```python
#Mars Hemisphere Data
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
hemisphere_image_urls = []

url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(url)

html = browser.html
soup = BeautifulSoup(html, 'html.parser')
tag = list(soup.find_all('h3'))

for x in range(0,4):
    browser.find_by_text(tag[x].text.strip()).click()
    

    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    image_url = soup.find('img', class_='wide-image')
    title = soup.find('h2', class_='title')
    

    hemisphere_image_urls.append({'title':title.text.strip(), 'image_url':f'http:/{image_url["src"]}'})
    
    browser.back()

print(hemisphere_image_urls)
```
